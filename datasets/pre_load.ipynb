```python
precent_of_data = 0.001
```

```python
import os
import pandas as pd
import numpy as np
from tqdm import tqdm
path = "./data/taxi_log_2008_by_id"
os.chdir(path)
files = os.listdir()

total_files = len(files[1:])
target_files = int(total_files * precent_of_data)

gps_data=pd.read_csv(files[0],names=['user_id','datetime','lat','lng'])

for i,file in enumerate(tqdm(files[1:], desc="Processing files")):
    tmp = pd.read_csv(file, names=['user_id', 'datetime', 'lng', 'lat'])
    if not tmp.empty:
        gps_data = pd.concat([gps_data, tmp])
    if i + 1 >= target_files:
        break
 
gps_data
```

```python
gps_data1=gps_data.sort_values(by=['user_id','datetime'],ignore_index=True)
gps_data1.drop_duplicates(inplace=True,ignore_index=True)
gps_data1['datetime'] = pd.to_datetime(gps_data1['datetime'], format='%Y-%m-%d %H:%M:%S')
gps_data1
```

```python
gps_data1=gps_data1[(gps_data1['lat']>39.4)&
                  (gps_data1['lat']<41.6)&
                  (gps_data1['lng']>115.7)&
                  (gps_data1['lng']<117.4)]
gps_data1
```

```python
gps_data1 = gps_data1.copy()
gps_data1['poi_id'] = gps_data1.groupby(['lng', 'lat']).ngroup()
gps_data1
```

```python
raw_df = gps_data1.copy()
raw_df.drop(columns=['lng', 'lat'], inplace=True)
raw_df = raw_df[['user_id', 'poi_id', 'datetime']]
raw_df
```

```python
coor_df = gps_data1[["poi_id", "lat", "lng"]].drop_duplicates()
coor_df.set_index('poi_id', inplace=True)
coor_df = coor_df.sort_index()
coor_df
```

```python
coor_df['lng'] = coor_df['lng'].transform(
    lambda x: (x - x.min()) / (x.max() - x.min())
)
coor_df['lat'] = coor_df['lat'].transform(
    lambda x: (x - x.min()) / (x.max() - x.min())
)
coor_df
```

```python
os.chdir("../../")
raw_df.to_hdf('taxi.h5', key='data', mode='w')
coor_df.to_hdf('taxi.h5', key='poi', mode='a')
```
